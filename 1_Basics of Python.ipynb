### üìñControl Flows

Control flow is the order in which individual statements, instructions, or function calls are executed or evaluated. The control flow of a Python program is regulated by conditional statements, loops, and function calls.

Here are a few examples of **conditional statements**: 

```python
x = 2 #change this value and see the results
if x >= 0:
    print ('x is positive')
else:
    print ('x is negative')
```

```python
x= -5
if x>0 and x<10:
    print ("x is between 1..9")
elif x<20:
    print ("x is between 10..19")
elif x<=0:
    print ("x is less or equal than 0")
else:
    print ("x is greater or equal than 20")
```

```python
# let's create a list of odd and even numbers <100
odd_list=[]
even_list=[]

for x in range(0,100):
    if x % 2 == 0: # % = modulos
        even_list.append(x)
        continue #restart loop
    odd_list.append(x)
print ("list of odd numbers", odd_list)
print ("================================================")
print ("list of even numbers", even_list)
```

A more complex example of guessing a random number using **control flows**:

```python
import random

no_attempts = 0
target_num = random.randrange(0, 100)

while True:
    no_attempts += 1
    num_input = input("Enter a valid number between 0 and 100: ")
    
    try:
        num = int(num_input)
    except ValueError:
        print("Please enter a valid integer.")
        continue
    
    if num < target_num:
        print("Your guess is smaller than the target number.")
    elif num > target_num:
        print("Your guess is greater than the target number.")
    else:
        print("Right answer :-)")
        print("Attempts needed: " + str(no_attempts))
        break
```

### üìñError Handing

Exceptions allow you to prevent the program from crashing even when an error occurs:

```python
# Try to convert the string "Test" to an integer (this will raise an error)
try:    
    int("Test")
# If an error occurs in the try block, catch the exception
except Exception as err:
    # Print out a custom error message along with the system-generated error
    print(("Error occured: ", err))
    # Terminate the script execution after encountering the error
    exit()
```

```python
# Start an infinite loop
while True:  
    try:
        x = int(input("Please enter a number: "))
        break 
    # This block will be executed if the conversion above fails.
    except ValueError:  
        print("Oops!  That was no valid number.  Try again...")
```

```python
try:
    outfile = open("x:/tmp/output.txt", "w")
    outfile.write("any output to file\n")
    outfile.close()
    print ("File successfully written.")

except Exception as err:
    print(("Could not write file: ", err))
    exit()
```

### üìñWorking with Time

```python
import time
# Print the current time in seconds since the epoch (January 1, 1970, UTC)
print (time.time())
```

```python
import time

# Record the current time before the loop starts
start = time.time()

# Run a loop from 1 to 999999 (it does nothing, but serves to consume time)
for i in range(1,1000000):
    pass

# Record the current time after the loop ends
end = time.time()

runtime = end - start
print ("Process took %.3f sec." % (runtime))
```

```python
localtime = time.localtime(time.time())
print ("Year: ", localtime.tm_year)
print ("Month: ", localtime.tm_mon)
print ("Day: ", localtime.tm_mday)
```
